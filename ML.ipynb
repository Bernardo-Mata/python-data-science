<h2>Regresion Logistica<h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("dataFrames/Machine Learning/datos_seguro.csv")
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['edad']], df['compra'], test_size=0.9)
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
modelo.score(X_test, y_test)
```

```python
datos_nuevos = pd.DataFrame({
    'edad': [25, 35, 45, 55]
})
```

```python
datos_nuevos
```

```python
probabilidades = modelo.predict_proba(datos_nuevos)
probabilidades
```

```python
prob_compra = probabilidades[:, 1]
prob_compra
```

```python
plt.scatter(df.edad, df.compra, color='red', marker='+')
plt.plot(datos_nuevos.edad, prob_compra, color='blue', marker='o')
```

```python
df
```

```python
plt.scatter(df.edad, df.compra, color='red', marker='+')
```

```python
""" 
Vamos a proceder a separar estos datos en conjuntos de entrenamiento y prueba,
 y luego a evaluar nuestro modelo de regresión logística.

Utiliza la función correcta para dividir tu DataFrame en cuatro partes: X_entrena, X_prueba, y_entrena, y_prueba.
 Asegúrate de que la proporción de entrenamiento sea del 90% del total de los datos.

Crea la variable modelo donde almacenes el modelo de regresión logística usando la clase adecuada de
 sklearn y entrénalo con los conjuntos X_entrena y y_entrena.

Evalúa el desempeño de tu modelo con los datos de entrenamiento usando el método correcto para este fin.
"""

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df[['edad', 'salario']]
y = df['compra']

X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, test_size=0.2)

```

```python
modelo = LogisticRegression()
modelo.fit(X_entrena, y_entrena)
```

```python
# Usar las variables correctas creadas por train_test_split
# en este notebook se llaman X_prueba y y_prueba (o X_test/y_test según la celda),
# así que usamos las que existen: X_prueba y y_prueba
modelo.score(X_prueba, y_prueba)
```

```python
datos_nuevos = pd.DataFrame({
    'edad': [25, 35, 45, 55, 60, 75],
    'salario': [23000, 32000, 43000, 57000, 68000, 78000]
})
```

```python
proba = modelo.predict_proba(datos_nuevos)
proba
```

```python
plt.scatter(df['edad'], df['compra'])
plt.plot(datos_nuevos['edad'], proba[:, 1], color='red', marker='o')
```

<h2>Arboles de desicion<h2>

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.species.unique()
```

```python
#transformar variable categórica a numérica
le = LabelEncoder()
species = le.fit_transform(df['species'])
```

```python
#variables independientes
X = df.drop('species', axis=1)
```

```python
#variables dependiente
y = species
```

```python
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, test_size=0.8, random_state=42)
```

```python
arbol = DecisionTreeClassifier()
```

```python
arbol.fit(X_entrena, y_entrena)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(12,8))
plot_tree(decision_tree=arbol, filled=True, feature_names=df.columns.tolist(), class_names=['setosa', 'versicolor', 'virginica'])
```

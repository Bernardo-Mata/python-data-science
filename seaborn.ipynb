<h2>Graficos de nivel inferior</h2>

```python
g = sns.PairGrid(pinguins, hue="species");
```

```python
g = sns.PairGrid(pinguins, hue="species", corner=True);
```

```python
g = sns.PairGrid(pinguins, hue="species", corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level= 5,
           color=".2");
```

```python
g = sns.PairGrid(pinguins, hue="species", corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level= 5,
           color=".2");
g.map_lower(sns.scatterplot,
           marker="+");
```

```python
g = sns.PairGrid(pinguins, hue="species", corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level= 5,
           color=".2");
g.map_lower(sns.scatterplot,
           marker="+");
g.map_diag(sns.histplot, element="step", linewidth=0,kde=True);
```

```python
g = sns.PairGrid(pinguins, hue="species", corner=True);
g.map_lower(sns.kdeplot,
           hue=None,
           level= 5,
           color=".2");
g.map_lower(sns.scatterplot,
           marker="+");
g.map_diag(sns.histplot, element="step", linewidth=0,kde=True);
g.add_legend(frameon=True, title="Especies de pinguins");
```
